```python
#importing the libraries
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#reading the data 
movies = pd.read_csv("my_movies.csv")
movies
```

```python
movies.shape
```

```python
movies.info()
```

```python
#Pre processing 
```

as the data is not in transation formation so we are using the transaction Encoder to convert the data into binary

```python
movies1=movies.drop(columns=['V1','V2','V3','V4','V5'])
movies1
```

```python
#applying the APRIORI Algorithm
```

```python
frequent_itemsets = apriori(movies1,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
rules 
rules.sort_values("lift",ascending=False)
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'],rules['confidence'],alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules['support'],rules['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```
