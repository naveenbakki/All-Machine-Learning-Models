# Random Forest for  Company Dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#loading the data
company = pd.read_csv("Company_Data.csv")
company
```

```python
company.shape
```

```python
company.info()
```

```python
#converting the continuous data to categorical data of target variable
company.loc[company.Sales >=8.5 , 'Revised_Sales'] = 'High'
company.loc[company.Sales <8.5, 'Revised_Sales'] = 'Low'
```

```python
company
```

```python
company1=company.drop(['Sales' ],axis=1)
```

```python
company1
```

```python
#converting all the categorical data to numeric by applying the labelEncoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
#applying label encoder to ShelveLoc,Urban,US columns
company1['ShelveLoc'] = label_encoder.fit_transform(company1['ShelveLoc'])
company1['Urban'] = label_encoder.fit_transform(company1['Urban'])
company1['US'] = label_encoder.fit_transform(company1["US"])
```

```python
company1
```

```python
company1.Revised_Sales.value_counts()
```

```python
#dividing the data into x==>features , y==>target variable
```

```python
x = company1.drop("Revised_Sales" , axis=1)
y = company1.Revised_Sales
```

```python
#features
x
```

```python
#target variable
y
```

```python
#creating the random forest classification model
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#creating the model
num_trees = 100
max_features = 6
kfold = KFold(n_splits=5)

model = RandomForestClassifier(n_estimators=num_trees,max_samples=0.7,max_features=max_features,random_state=10)
results = cross_val_score(model,x,y,cv=kfold)
print(results.mean())
```

```python
model.fit(x,y)
```

```python
model.estimators_
```

```python
company1["model_rf_pred"]  = model.predict(x)
```
